```python
import glob, os
import json
import pandas as pd
from area import area
```

```python
root=os.getcwd()
```

```python
directories=[
            "dengue-cases-central",
            "dengue-cases-north-east",
            "dengue-cases-south-east",
            "dengue-cases-south-west",
            ]
```

```python
df_comb=pd.DataFrame()
```

```python
#if you encounter error, restart kernel.
for directory in directories:
    os.chdir(root+f'/../assets/datagov/{directory}/')
    current_list=[]
    for file in glob.glob("*.geojson"):
        js=json.load(open(file))
        features=js['features']
#         current_list.extend(features)
        df=pd.DataFrame(features)
        df['file']=directory
        df['desc']=df['properties'].apply(lambda x: x['Description'])
        df['coords']=df['geometry'].apply(lambda x: x['coordinates'][0])
        df.drop(columns=["properties","type"],inplace=True)
#         display(df)
        df_comb=pd.concat([df_comb,df],axis=0,ignore_index=True)
```

```python
df_comb
```

```python
regex=r'JOIN_COUNT<\/th>\s?<td>(\d+)<\/td>'
df_comb['breeding_habitat_count']=df_comb['desc'].str.extract(pat=regex)

df_comb['area']=df_comb['geometry'].apply(lambda x: area(x))
```

```python
df_comb.drop(columns=["desc","geometry"],inplace=True)
```
